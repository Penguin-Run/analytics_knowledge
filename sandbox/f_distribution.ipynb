## F-distribution

### Basic

```python
sample = np.random.f(5, 2, 100000)
```

```python
plt.figure(figsize=(15, 5))
(
    sns.histplot(x = sample[sample < 5], bins=100)
)
```

```python
sample = np.random.f(100, 100, 100000)
```

```python
plt.figure(figsize=(15, 5))
(
    sns.histplot(x = sample[sample < 10], bins=100)
)
```

### Modelling

#### From chi-squared

```python
n = 100000
a = np.random.chisquare(5, n)
b = np.random.chisquare(2, n)

sample = a / b
```

```python
plt.figure(figsize=(15, 5))
(
    sns.histplot(x = sample[sample < 10], bins=100)
)
```

equal to f which is great

#### From normal with abs (try)

```python
n = 10000000
a = abs(np.random.standard_normal(n))
b = abs(np.random.standard_normal(n))

sample = a / b
```

```python
plt.figure(figsize=(15, 5))
(
    sns.histplot(x = sample[sample < 10], bins=100)
)
```

```python
np.percentile(sample, 50)
```
